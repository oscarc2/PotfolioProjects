```python
#import relevant libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# read in the data

df = pd.read_csv(r'C:\Users\ocoro\Downloads\movies.csv')

df #take a look at the data

```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))

#some missing values therefore we will just remove here

df = df.dropna()

#df #take a look at the data
```

```python


#some missing values therefore we will just remove here for simplicity

df = df.dropna()

#Check again for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))

```

```python


# Data Types for the columns

print(df.dtypes)



     
```

```python
#remove decimal point by changing data type to int64

df['votes'] =df['votes'].astype('int64')

df['budget'] =df['budget'].astype('int64')

df['gross'] =df['gross'].astype('int64')

df['runtime'] =df['runtime'].astype('int64')

print(df.dtypes)
```

```python
df #take a look at the data
```

```python
#discrepency in year and released columns so exctracting year from correct released column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python


# Order our Data a little bit to see

df.sort_values(by=['gross'], inplace=False, ascending=False)



     
```

```python
#scatter plot - budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])
plt.title ('Budget vs Gross Earnings')
plt.xlabel ('Budget for Film ')
plt.ylabel ('Gross Earnings')

plt.ticklabel_format(axis='both', style='sci', scilimits=(0,0))
plt.xlim([0,4e8])
plt.ylim([0,4e9])

plt.show()
```

```python
#Plot budget vs gross in regression

sns.regplot(x = 'budget', y = 'gross', data = df, line_kws = {'color' : 'blue'})
```

```python
# looking at any correlations

df.corr(method = 'pearson')
```

```python
#visual representation of correlations



correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix')

plt.xlabel('Categories')

plt.ylabel('Categories')

plt.show()



     
```

```python
# Looking at the top 15 compaies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python

sns.stripplot(x="rating", y="gross", data=df)
plt.title ('Gross Revenue and Ratings')
```

```python
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix')

plt.xlabel('Categories')

plt.ylabel('Categories')

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python


#Looking at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)



     
```
